## 1️⃣ Cargar el dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('GoogleApps.csv')
print('Dataset cargado correctamente ✅')
print(f'Filas: {len(df)} | Columnas: {len(df.columns)}')
```

## 3️⃣ Preguntas simples

```python
print('Promedio de Rating:   ', round(df['Rating'].mean(), 4))
print('Mediana de Installs:  ', df['Installs'].median())
print('Precio máximo:        ', df['Price'].max())
print('Tamaño mínimo (MB):   ', df['Size'].min())
```

## 4️⃣ Filtrado — Una condición
> Seleccionar solo las filas que cumplen una condición.

```python
# Apps con Rating mayor a 4.9
apps_altas = df[df['Rating'] > 4.9]
print(f'Apps con Rating > 4.9: {len(apps_altas)}')
apps_altas.head()
```

```python
# Promedio de instalaciones de apps con Rating > 4.9
promedio = df[df['Rating'] > 4.9]['Installs'].mean()
print(f'Promedio de Installs (Rating > 4.9): {promedio:,.0f}')
```

## 5️⃣ Filtrado — Dos condiciones
> Usamos `&` para AND y `|` para OR. Cada condición va entre paréntesis `()`.

```python
# Apps GRATUITAS con Rating > 4.9
filtro = df[(df['Rating'] > 4.9) & (df['Type'] == 'Free')]
print(f'Apps gratuitas con Rating > 4.9: {len(filtro)}')
filtro.head()
```

```python
# Promedio de instalaciones de apps gratuitas con Rating > 4.9
resultado = df[(df['Rating'] > 4.9) & (df['Type'] == 'Free')]['Installs'].mean()
print(f'Promedio de Installs (Gratuitas + Rating > 4.9): {resultado:,.0f}')
```

## 6️⃣ Visualización con Matplotlib

```python
# Gráfica 1: Promedio de Rating — Gratuitas vs De Pago
promedios = df.groupby('Type')['Rating'].mean()

plt.figure(figsize=(6, 4))
promedios.plot(kind='bar', color=['steelblue', 'orange'], edgecolor='white')
plt.title('Promedio de Rating: Gratuitas vs De Pago')
plt.xlabel('Tipo de app')
plt.ylabel('Rating promedio')
plt.xticks(rotation=0)
plt.ylim(4.0, 4.4)
plt.tight_layout()
plt.show()
```

```python
# Gráfica 2: Top 10 categorías con más apps
top_categorias = df['Category'].value_counts().head(10)

plt.figure(figsize=(10, 5))
top_categorias.plot(kind='barh', color='steelblue', edgecolor='white')
plt.title('Top 10 Categorías con más Apps')
plt.xlabel('Número de apps')
plt.ylabel('Categoría')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Gráfica 3: Distribución de Ratings
plt.figure(figsize=(8, 4))
df['Rating'].plot(kind='hist', bins=20, color='steelblue', edgecolor='white')
plt.title('Distribución de Ratings de las Apps')
plt.xlabel('Rating')
plt.ylabel('Número de apps')
plt.tight_layout()
plt.show()
```
